### Q12: Which product categories, SKUs, or regions contribute the most to total revenue? Perform a Pareto analysis (80/20 rule).

Identify top performers and determine if 80% of revenue comes from 20% of products/regions.

```python
import pandas as pd
df = pd.read_csv("cleaned_product_sales_dataset.csv")

product_revenue = (
    df.groupby("Product_Name")["Revenue"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

product_revenue["Cumulative_Revenue"] = product_revenue["Revenue"].cumsum()
total_revenue = product_revenue["Revenue"].sum()

product_revenue["Cumulative_Percentage"] = (
    product_revenue["Cumulative_Revenue"] / total_revenue * 100
)

top_products_80 = product_revenue[
    product_revenue["Cumulative_Percentage"] <= 80
]

percentage_products = (
    len(top_products_80) / len(product_revenue) * 100
)

percentage_products
```

### Q13: Calculate and analyze key performance indicators (KPIs) such as Average Order Value (AOV), order frequency, customer lifetime value indicators, and profit margins.

Define and compute relevant business metrics from the data.

```python
"""Calculated KPIs like AOV, order frequency, CLV proxy, and profit margin to evaluate business performance."""
import pandas as pd

df = pd.read_csv("cleaned_product_sales_dataset.csv")
df['Order_Date'] = pd.to_datetime(df['Order_Date'])




total_revenue = df['Revenue'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order_ID'].nunique()
total_customers = df['Customer_Name'].nunique()
average_order_value = total_revenue / total_orders
order_frequency = total_orders / total_customers
customer_lifetime_value = average_order_value * order_frequency

profit_margin = (total_profit / total_revenue) * 100

average_order_value, order_frequency, customer_lifetime_value, profit_margin
```

### Q14: Identify customer segments based on purchasing behavior (e.g., high-value vs. low-value customers, frequent vs. occasional buyers). How many segments exist?

Segment customers using clustering or quartile-based analysis.

```python
import pandas as pd

# Load cleaned dataset
df = pd.read_csv("cleaned_product_sales_dataset.csv")

# Create customer-level metrics
customer_metrics = (
    df.groupby("Customer_Name")
    .agg(
        total_revenue=("Revenue", "sum"),
        order_count=("Order_ID", "nunique")
    )
    .reset_index()
)

# Revenue-based segmentation using quartiles
customer_metrics["Revenue_Segment"] = pd.qcut(
    customer_metrics["total_revenue"],
    q=4,
    labels=["Low Value", "Mid-Low Value", "Mid-High Value", "High Value"]
)

# Frequency-based segmentation using custom bins
customer_metrics["Frequency_Segment"] = pd.cut(
    customer_metrics["order_count"],
    bins=[0, 1, 2, 5, customer_metrics["order_count"].max()],
    labels=["Occasional", "Low-Frequent", "Frequent", "Very Frequent"],
    include_lowest=True
)

# Combine segments
customer_metrics["Customer_Segment"] = (
    customer_metrics["Revenue_Segment"].astype(str)
    + " & "
    + customer_metrics["Frequency_Segment"].astype(str)
)

# Number of unique segments
customer_metrics["Customer_Segment"].nunique()
```

### Q16: Create a multi-panel visualization dashboard showing: (1) Revenue trend, (2) Top 10 products/categories, (3) Sales distribution, (4) Customer segments.

Build a professional visualization with multiple subplots.

```python
"""Built a multi-panel dashboard to visualize revenue trends, top products, sales distribution, and customer segments."""
import pandas as pd
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_csv("cleaned_product_sales_dataset.csv")
df["Order_Date"] = pd.to_datetime(df["Order_Date"])

# -------------------------------
# Prepare data for plots
# -------------------------------

# 1. Revenue trend (monthly)
monthly_revenue = (
    df.groupby(df["Order_Date"].dt.to_period("M"))["Revenue"]
    .sum()
)
monthly_revenue.index = monthly_revenue.index.to_timestamp()

# 2. Top 10 products by revenue
top_products = (
    df.groupby("Product_Name")["Revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# 3. Customer segmentation
customer_revenue = df.groupby("Customer_Name")["Revenue"].sum()
q1 = customer_revenue.quantile(0.25)
q3 = customer_revenue.quantile(0.75)

customer_segments = customer_revenue.apply(
    lambda x: "High Value" if x > q3 else
              "Mid Value" if x > q1 else
              "Low Value"
)

segment_counts = customer_segments.value_counts()

# -------------------------------
# Create multi-panel dashboard
# -------------------------------

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Revenue trend
axes[0, 0].plot(monthly_revenue)
axes[0, 0].set_title("Monthly Revenue Trend")
axes[0, 0].set_xlabel("Month")
axes[0, 0].set_ylabel("Revenue")

# Plot 2: Top 10 products
top_products.plot(kind="bar", ax=axes[0, 1])
axes[0, 1].set_title("Top 10 Products by Revenue")
axes[0, 1].set_xlabel("Product")
axes[0, 1].set_ylabel("Revenue")
axes[0, 1].tick_params(axis="x", rotation=45)

# Plot 3: Revenue distribution
axes[1, 0].hist(df["Revenue"], bins=30)
axes[1, 0].set_title("Revenue Distribution")
axes[1, 0].set_xlabel("Revenue")
axes[1, 0].set_ylabel("Frequency")

# Plot 4: Customer segments
segment_counts.plot(kind="bar", ax=axes[1, 1])
axes[1, 1].set_title("Customer Segments Distribution")
axes[1, 1].set_xlabel("Customer Segment")
axes[1, 1].set_ylabel("Number of Customers")

plt.tight_layout()
plt.show()
```

### Q21: Perform cohort analysis to understand customer retention and repeat purchase behavior over time. Identify which cohorts have the highest lifetime value and why.

Create cohort tables based on first purchase date and analyze purchasing patterns across cohorts to identify trends in customer loyalty.

```python
"""Performed cohort analysis based on first purchase date to study customer retention over time."""
import pandas as pd
df = pd.read_csv("cleaned_product_sales_dataset.csv")
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df["Order_Month"] = df["Order_Date"].dt.to_period("M")

df["Cohort_Month"] = (
    df.groupby("Customer_Name")["Order_Month"]
    .transform("min")
)

df["Cohort_Index"] = (
    (df["Order_Month"].dt.year - df["Cohort_Month"].dt.year) * 12 +
    (df["Order_Month"].dt.month - df["Cohort_Month"].dt.month) + 1
)
cohort_table = (
    df.groupby(["Cohort_Month", "Cohort_Index"])["Customer_Name"]
    .nunique()
    .reset_index()
)

cohort_pivot = cohort_table.pivot(
    index="Cohort_Month",
    columns="Cohort_Index",
    values="Customer_Name"
)

cohort_pivot
```

### Q23: Calculate and analyze the Price Elasticity of Demand for different product categories. How sensitive is demand to price changes? Provide actionable pricing recommendations.

Use statistical regression or correlation analysis to determine price elasticity and suggest optimal pricing strategies for revenue maximization.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("cleaned_product_sales_dataset.csv")

df = df[(df["Unit_Price"] > 0) & (df["Quantity"] > 0)]

df["log_price"] = np.log(df["Unit_Price"])
df["log_quantity"] = np.log(df["Quantity"])

price_elasticity = {}

for category in df["Category"].unique():
    data = df[df["Category"] == category]
    if len(data) > 2:
        slope = np.polyfit(data["log_price"], data["log_quantity"], 1)[0]
        price_elasticity[category] = slope

price_elasticity

```
